# removing the less data countries

```python
def shorten_categories(categories,cutoff):
    categorical_map={}
    for i in range(len(categories)):
        if categories.values[i]>=cutoff:
            categorical_map[categories.index[i]]=categories.index[i]
        else:
            categorical_map[categories.index[i]]="Other"
    return categorical_map        
```

```python
country_map=shorten_categories(categories,400)
cf["Country"]=cf["Country"].map(country_map)
print(cf["Country"].value_counts())
```

```python
x=cf["YearsCodePro"].unique()
```

```python
def clean_experience(x):
    if x ==  'More than 50 years':
        return 50
    if x ==  'Less than 1 year':
        return 0.5
    return float(x)

cf['YearsCodePro'] = cf['YearsCodePro'].apply(clean_experience) 
```

```python
cf["YearsCodePro"].unique()
```

```python

y=cf["EdLevel"].unique()
```

```python
def clean_education(y):
    if 'Bachelor’s degree' in y:
        return 'Bachelor’s degree'
    if 'Master’s degree' in y:
        return 'Master’s degree'
    if 'Professional degree' in y or 'Other doctoral' in y:
        return 'Post grad'
    return 'Less than a Bachelors'

cf['EdLevel'] = cf['EdLevel'].apply(clean_education)
```

```python

cf["EdLevel"].unique()
```

# transform into machine understandable

```python
from sklearn.preprocessing import LabelEncoder
le_education=LabelEncoder()
cf['EdLevel']=le_education.fit_transform(cf['EdLevel'])
cf['EdLevel'].unique()
                                
```

```python
le_country=LabelEncoder()
cf["Country"]=le_country.fit_transform(cf["Country"])
cf["Country"].unique()
```

```python
x=cf.drop("Salary",axis=1)
y=cf["Salary"]
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
y_pred=regressor.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
import numpy as np
error=np.sqrt(mean_squared_error(y_test,y_pred))
r2=r2_score(y_test,y_pred)
print(error)
print(r2)
              
```

```python
from sklearn.tree import DecisionTreeRegressor
tree = DecisionTreeRegressor(random_state=0)
tree.fit(x_train, y_train)
y_pred=tree.predict(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
forest=RandomForestClassifier(n_estimators=100)
forest.fit(x_train,y_train)
y_pred=forest.predict(x_test)
```

```python
from sklearn.model_selection import GridSearchCV

max_depth = [None, 2,4,6,8,10,12]
parameters = {"max_depth": max_depth}

regressor = DecisionTreeRegressor(random_state=0)
gs = GridSearchCV(regressor, parameters, scoring='neg_mean_squared_error')
gs.fit(x_train, y_train)
```

```python

regressor = gs.best_estimator_

regressor.fit(x_train, y_train)
y_pred = regressor.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
import numpy as np
error=np.sqrt(mean_squared_error(y_test,y_pred))
r2=r2_score(y_test,y_pred)
print(error)
print(r2)
```
